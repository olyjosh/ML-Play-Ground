### Preproessing

```python
# df['experience_level'].nunique()

# Binary encoding: employment_type
df['employment_type'] = df['employment_type'].apply(lambda x: 1 if x == 'FT' else 0)

df
```

```python
df.experience_level.info()
```

```python
# One hot encoding: experience_level, job_title, employee_residence, company_location, company_size
df = df.join(pd.get_dummies(df['experience_level'], prefix='experience_level')).drop('experience_level', axis=1)
df.job_title =  df.job_title.factorize()[0] #df.join(pd.get_dummies(df['job_title'], prefix='job_title')).drop('job_title', axis=1)  # factorise instead
df.employee_residence =  df.employee_residence.factorize()[0] #df = df.join(pd.get_dummies(df['employee_residence'], prefix='employee_residence')).drop('employee_residence', axis=1) #factorise
df.company_location =  df.company_location.factorize()[0] #df = df.join(pd.get_dummies(df['company_location'], prefix='company_location')).drop('company_location', axis=1) # factorise
df = df.join(pd.get_dummies(df['company_size'], prefix='company_size')).drop('company_size', axis=1)

# Dropping salary (but using salary in USD), salary_currency
df = df.drop(['salary', 'salary_currency'], axis=1)

# df = df.map(lambda x : 1 if x == True else 0 if x == False else x)

df
```

### Modeling

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X, y = df.drop(['salary_in_usd'], axis=1), df['salary_in_usd']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


reg = RandomForestRegressor(n_jobs=-1)
reg.fit(X_train, y_train)
reg.score(X_test, y_test)
```
